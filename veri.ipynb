```python
import pandas as pd
import random

# Rastgele veri üretmek için sabitler
markalar = ["Samsung", "Apple", "Xiaomi", "Huawei", "Oppo", "OnePlus"]
modeller = {
    "Samsung": [("Galaxy S10", 500), ("Galaxy S20", 1000), ("Galaxy S21", 1500)],
    "Apple": [("iPhone 10", 1000), ("iPhone 11", 1500), ("iPhone 12", 2000)],
    "Xiaomi": [("Redmi Note 8", 300), ("Redmi Note 9", 500), ("Mi 11", 1000)],
    "Huawei": [("P30", 400), ("P40", 800), ("Mate 40", 1200)],
    "Oppo": [("A15", 200), ("Reno 5", 600), ("Find X3", 900)],
    "OnePlus": [("Nord", 500), ("8T", 1000), ("9 Pro", 1500)]
}
depolama_secenekleri = [64, 128, 256]
ram_secenekleri = [4, 6, 8, 12]

# Marka premium katsayıları
marka_premium = {
    "Samsung": 1.3,
    "Apple": 1.5,
    "Xiaomi": 0.9,
    "Huawei": 1.0,
    "Oppo": 0.8,
    "OnePlus": 1.2
}

def generate_data(n):
    data = []
    for _ in range(n):
        marka = random.choice(markalar)
        model, model_extra_price = random.choice(modeller[marka])
        depolama = random.choice(depolama_secenekleri)
        ram = random.choice(ram_secenekleri)
        yas = random.randint(0, 5)  # Telefonun yaşı (0-5 yıl arası)
        
        # Fiyat hesaplama
        base_price = 7000 * marka_premium[marka]  # Marka premium katsayısı
        age_penalty = 1000 * yas  # Yaşa bağlı fiyat düşüşü
        
        storage_price = depolama * 60  
        ram_price = ram * 120
        
        # Depolama ve RAM etkileşim katsayısı
        storage_ram_bonus = (depolama * ram) * 0.05
        
        # Popülarite etkisi
        popularity_factor = random.uniform(0.9, 1.2)
        
        # Vergi etkisi
        tax = base_price * 0.08
        
        fiyat = (
            base_price +
            model_extra_price +
            storage_price +
            ram_price +
            storage_ram_bonus -
            age_penalty +
            tax
        ) * popularity_factor
        
        fiyat = max(fiyat, 100)  # Minimum fiyat sınırı
        data.append([marka, model, depolama, ram, yas, round(fiyat, 2)])
    return data

def generate_random_data(n):
    data = []
    for _ in range(n):
        marka = random.choice(markalar)
        model, model_extra_price = random.choice(modeller[marka])
        depolama = random.choice(depolama_secenekleri)
        ram = random.choice(ram_secenekleri)
        yas = random.randint(0, 5)  # Telefonun yaşı (0-5 yıl arası)
        fiyat = random.randint(6000, 20000)  # Rastgele fiyat aralığı
        data.append([marka, model, depolama, ram, yas, round(fiyat, 2)])
    return data



columns = ["Marka", "Model", "Depolama", "RAM", "Yaş", "Fiyat"]
veri = generate_data(5000)
df = pd.DataFrame(veri, columns=columns)

df.to_csv("veri.csv", index=False)
```
